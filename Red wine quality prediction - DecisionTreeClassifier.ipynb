```python
# Import support libraries 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
# Load Dataset
data=pd.read_csv("winequality-red.csv")

data['citric acid'].replace(0,data['citric acid'].median(),inplace=True)
data['citric acid']=data['citric acid'].astype(float)
```

```python
mean=data.mean()
mean
```

```python
std_dev=data.std()
std_dev
```

```python
q1=mean-1.4*std_dev
```

```python
q3=mean+1.4*std_dev
```

```python
outlier=[(data<q1) | (data>q3)]
```

```python
corr_data=data[(data>q1) & (data<q3)]
```

```python
corr_data['fixed acidity'].fillna(corr_data['fixed acidity'].mean(),inplace=True)
corr_data['volatile acidity'].fillna(corr_data['volatile acidity'].mean(),inplace=True)
corr_data['citric acid'].fillna(corr_data['citric acid'].mean(),inplace=True)
corr_data['residual sugar'].fillna(corr_data['residual sugar'].mean(),inplace=True)
corr_data['chlorides'].fillna(corr_data['chlorides'].mean(),inplace=True)
corr_data['free sulfur dioxide'].fillna(corr_data['free sulfur dioxide'].mean(),inplace=True)
corr_data['total sulfur dioxide'].fillna(corr_data['total sulfur dioxide'].mean(),inplace=True)
corr_data['density'].fillna(corr_data['density'].mean(),inplace=True)
corr_data['pH'].fillna(corr_data['pH'].mean(),inplace=True)
corr_data['sulphates'].fillna(corr_data['sulphates'].mean(),inplace=True)
corr_data['alcohol'].fillna(corr_data['alcohol'].mean(),inplace=True)
```

```python
corr_data
```

```python
# Assign columns to variables
X=corr_data.drop(columns=['quality'])
y=data['quality']
```

```python
# Provide training size
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
# Choose model
model=DecisionTreeClassifier()
```

```python
# Give training to model
model=model.fit(X_train,y_train)
```

```python
# Model prections test value
predictions=model.predict(X_test)
```

```python
# Score check
score=accuracy_score(y_test,predictions)
```
