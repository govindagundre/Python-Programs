```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

month= ['Jan','Feb','Mar','Apr','May']
         
HYDsales=[50,60,55,75,65]
BANsales=[55,50,70,60,80]
index=np.arange(1,6)
width=0.30
plt.bar(index,HYDsales,width,color='green',label='HYD Sales')
plt.bar(index+width,BANsales,width,color='blue',label='BAN Sales')
plt.title('Horizaontally Stacked Bar Graph')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(index+width/2 ,month)
plt.legend(loc='best')
plt.show()
```

```python

month= ['Jan','Feb','Mar','Apr','May']
         
HYDsales=[50,60,55,75,65]
BANsales=[55,50,70,60,80]
index=np.arange(1,6)
#width=0.30
plt.bar(index,HYDsales,width,color='green',label='HYD Sales')
plt.bar(index,BANsales,width,color='blue',label='BAN Sales',bottom=HYDsales)
plt.title('Horizaontally Stacked Bar Graph')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(index+width/9 ,month)
plt.legend(loc='best')
plt.show()
```
